```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#labels = np.genfromtxt("labels.csv", names = ["image", "value", "type"])

labels = pd.read_csv("labels.csv", header = None, delim_whitespace=True, names = ["image", "value", "type"])

print(labels)
#print(labels.size)
#print(labels.shape)
```

```python
mono_count = 0
for row in range(labels.shape[0]):
    if labels.iloc[row]["type"] == "mono":
        mono_count += 1

print('take this figure and split by label. i.e. stacked bar chart')

poly_count = labels.shape[0] - mono_count


fig, ax = plt.subplots()

width = 0.35
ax.bar('Monocrystalline', mono_count, label='mono', width=width)
ax.bar('Polycrystalline', poly_count, label='poly', width=width)

plt.show()

# adjust width perhaps
```

```python
count_1_mono = 0
count_1_poly = 0

count_0_mono = 0
count_0_poly = 0

count_03_mono = 0
count_03_poly = 0

count_06_mono = 0
count_06_poly = 0

for row in range(labels.shape[0]):
    if labels.iloc[row]["value"] == 1.0:
        if labels.iloc[row]["type"] == "mono":
            count_1_mono += 1
        else:
            count_1_poly += 1
    elif labels.iloc[row]["value"] == 0.0:
        if labels.iloc[row]["type"] == "mono":
            count_0_mono += 1
        else:
            count_0_poly += 1
    elif labels.iloc[row]["value"] == 0.3333333333333333:
        if labels.iloc[row]["type"] == "mono":
            count_03_mono += 1
        else:
            count_03_poly += 1
    else:
        if labels.iloc[row]["type"] == "mono":
            count_06_mono += 1
        else:
            count_06_poly += 1

print("Shows distribution of values, 1, 0.6, 0.3, and 0", '\n')
print("maybe keep this but add labels above the bars. Then put this figure to the right and the one after this to the left")

width = 0.35

fig, ax = plt.subplots()

ax.bar(1.0, count_1_mono + count_1_poly, width)
ax.bar(0.6, count_06_mono + count_06_poly, width)
ax.bar(0.3, count_03_mono + count_03_poly, width)
ax.bar(0.0, count_0_mono + count_0_poly, width)

ax.set_ylabel('Count')
ax.set_title('Counts of types by label')

plt.show()
```

```python
#print(count_1_poly + count_06_poly + count_03_poly + count_0_poly)
#print(count_1_mono + count_06_mono + count_03_mono + count_0_mono)

labels = ["1.0", "0.6", "0.3", "0.0"]
mono_counts = [count_1_mono, count_06_mono, count_03_mono, count_0_mono]
poly_counts = [count_1_poly, count_06_poly, count_03_poly, count_0_poly]

width = 0.35
x = np.arange(len(labels))

fig, ax = plt.subplots()

rects1 = ax.bar(x - width/2, mono_counts, width, label='Monocrystalline')
rects2 = ax.bar(x + width/2, poly_counts, width, label='Polycrystalline')

ax.set_ylabel('Count')
ax.set_title('Counts of labels by type')
ax.set_xticks(x, labels)
ax.set(ylim=(0, 1100))
ax.legend(loc='upper left')

ax.bar_label(rects1, padding=1)
ax.bar_label(rects2, padding=1)

fig.tight_layout()
plt.show()
```
